```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import CSVs and convert each to a dataframe
cost_of_living = pd.read_csv('cost_of_living.csv')
ds_salaries = pd.read_csv('ds_salaries.csv')
levelsfyi_salaries = pd.read_csv('Levels_Fyi_Salary_Data.csv')
continents = pd.read_csv('Continents2.csv')
country_codes = pd.read_excel('country_codes.xlsx')
```

The cost of living plus rent index containts all the above, plus the cost of rent. This single index makes the Cost of Living, Rent, Groceries and Restaurant indicies redundant, so I'll remove them from the analysis and focus on the Cost of Living Plus Rent index going forward. 
Source: https://www.investopedia.com/ask/answers/100214/how-cost-living-index-calculated.asp#:~:text=A%20cost%20of%20living%20index%20compares%20the%20expenses%20from%20one,basic%20expenses%20rise%20over%20time.

The section below aims to answer the assignment prompt, which is to determine the top 5 places in the world where the salary of a data scientist will go the fartheset with respect to each index provided: Cost

Create a new data frame called cost data, split the city and the country columns, rank them based on cost of living plus rent index

```python
#Create a new data frame called cost data 
cost_data = cost_of_living

#Split the city and the country columns
cost_data['Country'] = cost_data['City'].str.split(',').str[-1].str.strip()
cost_data['City'] = cost_data['City'].str.split(',').str[0].str.strip()

#Rank cities based on cost of living plus rent index
cost_data = cost_data.sort_values('Cost of Living Plus Rent Index', ascending = False)
cost_data['Rank'] = cost_data.index + 1
```

Below I will join the relevant data sets together for additional analysis. The 'continents2' csv was downloaded from Kaggle and used as it has the continent name which I use later in my analysis.Link to the data can be found here: https://www.kaggle.com/datasets/andradaolteanu/country-mapping-iso-continent-region?resource=download 

```python
#Join the continents2 to cost data
cost_data = pd.merge(cost_data, continents, left_on = 'Country', right_on = 'name', how = 'left')

#Drop uneeded columns
drop_cols = ['name', 'country-code','iso_3166-2','intermediate-region', 'region-code']
cost_data.drop(columns = drop_cols, inplace = True)
```

```python
#Filter ds_salaries for data scientist salaries only, and find the average salary by country
data_salaries = ds_salaries[ds_salaries['job_title'] == 'Data Scientist']
data_salaries = data_salaries.groupby('employee_residence')['salary_in_usd'].mean().round(1)
```

```python
#Join data_salaries to cost_data, dropping countries where salaries aren't available as the prompt asks where your salary will go the farthest
cost_data = pd.merge(cost_data, data_salaries, left_on = 'alpha-2', right_on = 'employee_residence', how = 'inner')
```

```python
#Drop Uneeded Columns
drop_cols = ['alpha-3', 'intermediate-region-code','sub-region-code']
cost_data.drop(columns = drop_cols, inplace = True)
cost_data
```

```python
#Rename salary column
cost_data.rename(columns={'salary_in_usd':'salary'}, inplace=True)
```

To answer the question on how far your salary will go, the code blocks below will create a salary index, using US salaries as the baseline.

```python
#Find the salary for US which will serve as the denominator to calculate the salary index
us_salary = cost_data[cost_data['Country'] == 'United States']['salary'].unique()[0]
```

```python
#Calculate the salary index
cost_data['salary_index'] = round(cost_data['salary'] / us_salary,2)
```

```python
#EDA, Create a chart of countries by salary index
cost_data.groupby('Country')['salary_index'].mean().round(1).sort_values(ascending = False). plot(kind = 'bar', figsize=(20, 6))
plt.xticks(rotation=45, ha='right',fontsize=8)
plt.xlabel('Country')
plt.ylabel('Data Scientist Salary Index')
plt.title('Data Scientist Salary Index, by Country')
plt.show()
```

To determine how far our money will go for each index, we will compare the salary index to the cost index

```python
cost_data
```

```python
# Compare cost indicies to salary index to find relative affordability scores
cost_data['cost_of_living_score'] = round(cost_data['salary_index'] / cost_data['Cost of Living Index'],4) * 100
cost_data['rent_score'] = round(cost_data['salary_index'] / cost_data['Rent Index'],4) * 100
cost_data['cost_of_living_plus_rent_score'] = round(cost_data['salary_index'] / cost_data['Cost of Living Plus Rent Index'],4) * 100
cost_data['groceries_score'] = round(cost_data['salary_index'] / cost_data['Groceries Index'],4) * 100
cost_data['restaurant_score'] = round(cost_data['salary_index'] / cost_data['Restaurant Price Index'],4) * 100
cost_data['purchasing_score'] = round(cost_data['salary_index'] / cost_data['Local Purchasing Power Index'],4) * 100

cost_data
```

```python
#The cities where your salary goes the furthest when compared to the cost of living index are: Algiers, ALgeria; El Paso US; Wichita, US; Little Rock, US; and Memphis, US
top_5_cost_of_living = cost_data.sort_values('cost_of_living_score', ascending = False).head(5)
top_5_cost_of_living
```

```python
#The cities where your salary goes the furthest when compared to the rent index are: Algiers, Algeria; Jammu, India; Agra, India; Curitbia, Brazil and Belo Horizonte, Brazil
top_5_rent_score = cost_data.sort_values('rent_score', ascending = False).head(5)
top_5_rent_score
```

```python
#The cities where your salary goes the furthest when compared to the cost of living plus rent index are: Algiers, Algeria; EL Paso, US; Wichita, US; Little Rock, US; and Akron, US
top_5_cost_of_living_plus_rent_score = cost_data.sort_values('cost_of_living_plus_rent_score', ascending = False).head(5)
top_5_cost_of_living_plus_rent_score
```

```python
#The cities where your salary goes the furthest when compared to the groceries index are: Algiers, Algeria; EL Paso, US; Wichita, US; Little Rock, US; and Akron, US
top_5_groceries_score = cost_data.sort_values('groceries_score', ascending = False).head(5)
top_5_groceries_score
```

```python
#The cities where your salary goes the furthest when compared to the restaurant index are: Algiers, Algeria; EL Paso, US; Akron, US; Lexington, US; and Fort Collins, US
top_5_restaurant_score = cost_data.sort_values('restaurant_score', ascending = False).head(5)
top_5_restaurant_score
```

```python
#The cities where your salary goes the furthest when compared to the purchasing index are: Lagos, Nigeria; Algiers, Algeria; Rio de Janeiro, Brazil; Belo Horizonte, Brazil; Cebu, Philippines
top_5_purchasing_score = cost_data.sort_values('purchasing_score', ascending = False).head(5)
top_5_purchasing_score
```

The code block below plots summary statistics using the dataframe we created

```python
#Find the avergae cost index by country
cost_by_country = cost_data.groupby('Country')['cost_index'].mean().round(1).sort_values(ascending = False)
```

```python
cost_by_country.plot(kind = 'bar', figsize=(20, 6))
plt.xticks(rotation=45, ha='right',fontsize=8)
plt.xlabel('Country')
plt.ylabel('Cost of Living Index with Rent')
plt.title('Cost of Living Index, Including Rent, by Country')
plt.show()
```
